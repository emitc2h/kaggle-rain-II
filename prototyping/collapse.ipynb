# Back to cleaning
---------------------------------------

```python
cf = {
    'Ref' : (1.01, 1.19, 0.86),
    'RefComposite' : (1.01, 1.15, 0.87),
    'RhoHV' : (1.00, 1.05, 0.97)
}

for var in ['Ref', 'RefComposite', 'RhoHV']:
    train[var] = cf[var][0]*train[var].fillna(value=train['{0}_5x5_50th'.format(var)])
    train[var] = cf[var][1]*train[var].fillna(value=train['{0}_5x5_10th'.format(var)])
    train[var] = cf[var][2]*train[var].fillna(value=train['{0}_5x5_90th'.format(var)])
    
    del train['{0}_5x5_10th'.format(var)]
    del train['{0}_5x5_50th'.format(var)]
    del train['{0}_5x5_90th'.format(var)]
```

```python
train['interval'] = train['minutes_past'].diff()
```

```python
group_begin_mask = train['Id'].diff() != 0
```

```python
group_end_mask = train['Id'].diff().shift(-1) != 0
```

```python
pd.set_option('mode.chained_assignment',None)
train['interval'][group_begin_mask] = train['minutes_past'][group_begin_mask]
```

```python
train['interval'][group_end_mask] = 60 + train['interval'][group_end_mask] - train['minutes_past'][group_end_mask]
```

```python
for var in ['Ref', 'RefComposite', 'RhoHV', 'Zdr', 'Kdp']:
    train['Int_{0}'.format(var)] = train['interval'] * train[var]
    del train[var]
```

```python
expected  = train[['Id', 'Expected', 'radardist_km']]
expected  = expected.groupby('Id', as_index=False).first()

variables = train[['Id', 'Int_Ref', 'Int_RefComposite', 'Int_RhoHV', 'Int_Zdr', 'Int_Kdp']]
variables = variables.groupby('Id', as_index=False).sum()
```

```python
for var in ['Int_Ref', 'Int_RefComposite', 'Int_RhoHV', 'Int_Zdr', 'Int_Kdp']:
    variables[var] = variables[var]/60.0
```

```python
train = pd.merge(variables, expected, on='Id')
train
```

```python
train.fillna(-1111.0)
train = mytools.clean_stuck_gauges(train, threshold=20)
train = train[train['Expected'] < 50.0]

sm = mytools.ScatterMatrix(train)
sm.add('radardist_km', 0, 20)
sm.add('Int_Ref', 0, 50)
sm.add('Int_RefComposite', 0, 50)
sm.add('Int_RhoHV', 0.8, 1.05)
sm.add('Int_Zdr', -2,3)
sm.add('Int_Kdp', -10, 10)
sm.add('Expected', 0, 10)
```
